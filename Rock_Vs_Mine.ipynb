# Sonar Rock Vs Mine Preiction using Python

**Binary Classification problem**

```python
import numpy as np
import pandas as pd
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
sonar_data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/content/Copy of sonar data.csv', header=None)
```

```python
sonar_data.head()
```

```python
sonar_data.shape
```

```python
sonar_data.columns
```

**R Represents Rock**

**M Represents Rock**

```python
sonar_data[60]
```

```python
sonar_data[60].value_counts()
```

```python
sonar_data[60] = sonar_data[60].map({'M':1,'R':0})
```

**Map 1 to Mine**

**Map 0 to Rock**

```python
sonar_data.describe()
```

```python
sonar_data.groupby(60).mean()
```

```python
# separating data and Labels
X = sonar_data.drop(columns=60, axis=1)
Y = sonar_data[60]
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.1, stratify=Y, random_state=1)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

### Model Evaluation

```python
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print(training_data_accuracy)
```

**Training Data Accuracy = 0.8342245989304813**

```python
#accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print(test_data_accuracy)
```
